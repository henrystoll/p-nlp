```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

df = pd.DataFrame({'Max Features':             [10000, 15000, 20000, 25000, 30000, 35000, 40000, 45000, 50000],
              'Validation Accuracy':    [0.533, 0.546, 0.555, 0.559, 0.562, 0.561, 0.562, 0.560, 0.562]})
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
sns.set_style('ticks')
sns.lineplot(data=df, x='Max Features', y='Validation Accuracy', ax=ax)
plt.axvline(x=30000, color='r', linestyle='--')
plt.tight_layout()
plt.savefig('../img/max_features.png')
```
